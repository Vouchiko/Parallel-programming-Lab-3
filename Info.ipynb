```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))

results = ['slurm-93935.out', 'slurm-93937.out', 'slurm-93938.out', 'slurm-93939.out', 'slurm-93940.out']
files = [1,2,4,6,8,10,12]
for i in results:
    sizes = []
    times = []
    with open(f"results/{i}", "r") as file:
        next(file)
        for line in file:
            parts = line.split()
            sizes.append(int(parts[0]))
            times.append(float(parts[1]))
    plt.plot(sizes, times, marker='o', label=f': {i}')

plt.xlabel('Size')
plt.ylabel('Time, sec')
plt.title('Time vs Size')
plt.grid()
plt.legend()
plt.show()
```
